```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
student_data = {
    "Name": ["Alice", "Bob", "Charlie", "David", "Eva"],
    "Age": [22, 23, 21, 24, 22],
    "Grade": [88, 76, 90, 85, 95],
    "Sex": ["F", "M", "M", "M", "F"],
    "Student Number": [1001, 1002, 1003, 1004, 1005],
    "Education Level": ["Senior", "Junior", "Senior", "Freshman", "Junior"],
    "BS Branch": [
        "Computer Science",
        "Electrical Engineering",
        "Mechanical Engineering",
        "Civil Engineering",
        "Computer Science"
    ]
}
```

```python
df = pd.DataFrame(student_data)
```

```python
df_csv = pd.read_csv("Student_data.csv")
gender_avg = df_csv.groupby("Sex")["Grade"].mean()

gender_avg.plot(kind="bar")
plt.xlabel("Sex")
plt.ylabel("Average Grade")
plt.title("Average Grade by Gender")
plt.show()
```
